### Total matches won by home team and away team

```python
#Let's see total matches won by team
def total_match_team_won(team):
    index = Points_table[Points_table['team']==team]['wins'].keys()[0]
    return Points_table[Points_table['team']==team]['wins'][index]
```

```python
team = 'Man United'
total_matches_won = total_match_team_won(team)
print("Total matches won by {} are {}".format(team,total_matches_won))
```

### Total matches lost by home team and away team

```python
#Let's see total matches lost by team
def total_match_team_lost(team):
    index = Points_table[Points_table['team']==team]['lossess'].keys()[0]
    return Points_table[Points_table['team']==team]['lossess'][index]
```

```python
def top5_players(team):
    top5_goal = players_performance[players_performance['Club']==team][['Name','Goals']].sort_values('Goals',ascending=False)
    plt.pie(list(top5_goal['Goals'].head()),labels=list(top5_goal['Name'].head()),autopct='%0.1f%%')
    plt.title("Top 5 {}'s players Goals percentages".format(team))
```

### Head to head Past 5 matches result between team1 and team2

```python
def matches_played(team1,team2):
    plt.barh(['Past 5 match won by {}'.format(team1),'Past 5 match won by {}'.format(team2),'Past 5 match drwan'],list(New_Matches[(New_Matches['HomeTeam']=='Everton') & (New_Matches['AwayTeam']=='Man United')][['Past_5_match_won_by_home','Past_5_match_won_by_away','Past_5_match_drwan']].head(1).values[0]))
    plt.title("Past 5 head to head result between {} and {}".format(team1,team2))
    plt.xlabel("Number of matches")
```

```python
matches_played('Everton','Man United')
```

### HST (Total shots on target made by home team) and  AST (Total shots on target made by away team)

```python
def Total_HST(team1):
    Match_played = New_Matches[(New_Matches['HomeTeam']==team1)]
    return Match_played['HST'].sum()

def Total_AST(team2):
    Match_played = New_Matches[(New_Matches['AwayTeam']==team2)]
    return Match_played['AST'].sum()
```

```python
team1 = 'Everton'
team2 = 'Man United'
print("Total shots on target made by {} are {}".format(team1,Total_HST(team1)))
print("Total shots on target made by {} are {}".format(team2,Total_AST(team2)))
```
